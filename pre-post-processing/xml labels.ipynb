# Get DataFrame from csv

```python
df = pd.read_csv(os.path.join(base_path, 'csv_final.csv'))
```

```python
array = {}
```

```python
for i in range(df.shape[0]):
    item = df.iloc[i]
    name = item['name']
    if not re.match('.*(.jpg)', name):
        name += '.jpg'
    array[name] = df.iloc[i, 1:].tolist()
```

```python
array
```

# Create XML files

```python
img_path = 'content/drive/My Drive/all_img_final'
```

```python
key_files = ['train_keys.txt', 'test_keys.txt', 'val_keys.txt']
xml_paths = ['train_labels', 'test_labels', 'val_labels']
```

```python
for i in range(len(key_files)):    
    keys = []
    with open(os.path.join(base_path, key_files[i]), 'r') as f:
        keys = f.read().split('\n')[:-1]
    for key in keys:
        [xmin, ymin, xmax, ymax, width, height, class_name] = array[key]
        writer = Writer(os.path.join(img_path, key), width, height)
        writer.addObject(class_name, xmin, ymin, xmax, ymax)
        writer.save(os.path.join(base_path + '\\' + xml_paths[i], key.split('.')[0] + '.xml'))
```
